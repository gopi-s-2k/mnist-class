```python
from utils import *
import matplotlib.pyplot as plt
import numpy as np
```

```python
mnist = load_mnist()
print(mnist.keys())
X,y = mnist["data"].to_numpy(),mnist["target"].to_numpy().astype(np.int8)
```

```python
#lets show some digits from mnist
fig,axs = plt.subplots(5,2,figsize=(10,20))
for i in range(10):
    u,v = i//2,i%2
    some_digit = X[i]
    some_digit_image = some_digit.reshape(28,28)
    axs[u][v].imshow(some_digit_image,cmap="binary")
    axs[u][v].set_title(f"the digit is - {y[i]}")
```
